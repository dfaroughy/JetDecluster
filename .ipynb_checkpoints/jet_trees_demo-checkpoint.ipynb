# Jet declustering trees:

Use ```get_particles``` to open HepMC collider event files and extract particles 4-momenta and PDG id. To cluster particles into jets, use ```pyjet```.

The main function of this module is  ``` decluster```:

The function takes as argument a previously clustered ```PseudoJet``` and outputs a ```Networkx``` graph where each node represents a subjet splitting ***j -> j_a j_b*** from the jet declustering procedure. The resulting topology of the graph is a binary tree with the clustered jet sitting at the root and the final state particles sitting at the leaves (nodes without descendants). Before declustering, ``` decluster``` internally re-clusters the jet constituents with the Cambridge/Aachen algorithm as default. The main attribute of each node is a ```Pseudojet``` of the mother subjet ***j***. The user has the option of including additional node attributes if necessary (see end of notebook). 

````python

with get_particles(data='events.hepmc', stop=1) as event:
    for particles in event:
 
        #...cluster jets:  
        sequence=cluster(particles, R=1., p=-1, ep=True)  # cluster using anti-kt (p=-1) with cone radius R=1
        jets=sequence.inclusive_jets(ptmin=20.)            
        
        for jet in jets:

            tree=decluster(jet, R=1., p=0)  # reclusters using c/a (p=0)

            if tree:
                
                # list of node indices:
                idx_nodes=list(nx.nodes(tree))

                # available node attributes:
                jet_label=nx.get_node_attributes(tree,'jet_label')     # to be specified by the user, Default:'None'  
                subjets=nx.get_node_attributes(tree,'pseudojet')       # Default: PseudoJet(pt,eta,phi,mass)
                primary=nx.get_node_attributes(tree,'primary_branch')  # True/False, specifies if splitting is primary
                features=nx.get_node_attributes(tree,'feature')        # to be specified by the user, Default:'None' 
                labels=nx.get_node_attributes(tree,'label')            # to be specified by the user, Default:'None' 

                # print node index and subjets for the leading jet:
                print('--------------------------------------------------------------------------------------')    
                print('{:<8} {:<10} {:<10} {:<10} {}'.format('node','label','feat.','primary', 'subjet '))
                print('--------------------------------------------------------------------------------------')

                for n in idx_nodes:
                    print( '{:<8} {:<10} {:<10} {:<10} {}'.format(str(n), str(labels[n]) , str(features[n]),str(primary[n]), subjets[n]) )

            break

````

# Visualizing the jet  trees:

Use ```draw_trees()``` for visualizing jet declustering trees.

The tree nodes correspond to the subjets splittings  ***j -> j_a j_b***. During a splitting, the outgoing edge towards the right carries the most momentum (subjet ***j_a***). The edge color code describe the flow of momentum (normalized to the total jet pT) from one subjet splitting to the next... Yellow represent soft subjets while dark blue/green represent harder subjets.


````python
with get_particles(data='events.hepmc', stop=3) as event:
        
    for N,particles in enumerate(event):
        sequence=cluster(particles, R=1., p=-1, ep=True)
        jets=sequence.inclusive_jets(ptmin=20.0)            

        trees=[]
        for n,jet in enumerate(jets):

            tree=decluster(jet,jet_label="event={}, jet={}".format(N,n))                 
            trees.append(tree)
            
        draw_trees(trees)
        if tree:
            jet_label=nx.get_node_attributes(tree,'jet_label') 
````

````python
tree=decluster(jet, directed=True)  
draw_trees([tree])
````

# Jet labels

Jet labels can be included via:

```decluster(jet, jet_label=<label>)```

The ```<label>``` information will be stored as a ```Networkx``` attribute in the root node of the tree and can be displayed when drawing the jet trees.

````python
with get_particles(data='events.hepmc', stop=3) as event:
        
    for N,particles in enumerate(event):
        sequence=cluster(particles, R=1., p=-1, ep=True)
        jets=sequence.inclusive_jets(ptmin=20.0)            

        trees=[]
        for n,jet in enumerate(jets):

            tree=decluster(jet,jet_label="event={}, jet={}".format(N,n))                 
            trees.append(tree)
            
        draw_trees(trees, jet_label=True)
````

# Substructure features and labels

In order to assign node features and/or labels useful for jet substructure studies, one can use ```node_feature``` and ```node_label```:

```decluster(jet, node_feature=<func_1>, node_label=<func_2>)```

The features and/or labels are computed at each subjet splitting from the input functions ```<func_1>```, ```<func_2>```, and added to each tree node as ```Networkx``` attributes. The inputs ```<func_1>```, ```<func_2>``` must be functions of the mother and daughters ```PseudoJet```, like in the example below. 

````python
from particle import Particle

# feature example:

def z(j,ja,jb):  # j -> ja jb, pt(ja) > pt(jb)
    z=min(ja.pt,jb.pt)/(ja.pt+jb.pt)
    return round(z,3)

# label example: this requires the 'particles' package for particle ID.

def get_particle_id(subjet):  # assigns particle ID to every final state particle (leaf nodes)
    if len(subjet)==1:
        return Particle.from_pdgid(subjet.pid)
    else:
        return None

    
# apply:
with get_particles(data='events.hepmc', stop=1) as event:
    for particles in event:
 
        sequence=cluster(particles, R=1., p=-1, ep=True)  # cluster using anti-kt
        jets=sequence.inclusive_jets(ptmin=20.0)            
        
        for jet in jets:
            
            tree=decluster(jet, node_feature=z, node_label=get_particle_id) 
            
            if tree:

                subjets=nx.get_node_attributes(tree,'pseudojet')        
                primary=nx.get_node_attributes(tree,'primary_branch')  
                features=nx.get_node_attributes(tree,'feature')         
                labels=nx.get_node_attributes(tree,'label')              

                # print node attributes for first jet:
                print('---------------------------------------------------------------------------------------------------')    
                print('{:<8} {:<10} {:<10} {:<10} {}'.format('node','PDG id',' z','primary', 'subjet '))
                print('---------------------------------------------------------------------------------------------------')

                for n in nx.nodes(tree):
                    print( '{:<8} {:<10} {:<10} {:<10} {}'.format(n,
                                                                  str(labels[n]), 
                                                                  str(features[n]),
                                                                  str(primary[n]),
                                                                  subjets[n]) )                
            break
            
````

# Example: Soft Drop Groomer

We implement from scratch the soft drop jet groomer. We display the first five groomed jets vs the non-groomed jets.

````python
# Soft-Drop implemention:

R=1.0

def SoftDrop(j, ja, jb, zcut=0.3, beta=1.0, R0=1.0):
    Delta=deltaR(ja,jb)
    softdrop=( min(ja.pt,jb.pt)/(ja.pt+jb.pt) > zcut * pow(Delta/R0,beta) )
    return softdrop


def jetlabel(jet):
    return jet


with get_particles(data='events.hepmc', stop=2) as event:
    
    for particles in event:
 
        #...cluster jets:  
        sequence=cluster(particles, R=R, p=-1, ep=True)  # cluster using anti-kt
        jets=sequence.inclusive_jets(ptmin=20.0)            
        
        for jet in jets:

            tree=decluster(jet, node_feature=SoftDrop, jet_label=r'jet ($p_t$={} GeV)'.format(round(jet.pt,1)), R=R)
            tree_=tree.copy() # exact copy of tree to-be-groomed  

            if tree_:
                
                subjets=nx.get_node_attributes(tree_,'pseudojet')  
                primary=nx.get_node_attributes(tree_,'primary_branch') 
                features=nx.get_node_attributes(tree_,'feature')  

                idx_drop=[]

                for idx, softdrop in nx.get_node_attributes(tree_,'feature').items():
                    if primary[idx]:
                        if softdrop:
                            groom_idx=idx # get truncation node
                            break
                        else:
                            idx_drop.append(idx) # collect node that will be (soft)dropped
                
                for node in idx_drop:
                    tree_.remove_node(node)

                tree_groomed=nx.relabel_nodes(tree_.subgraph(max(nx.connected_components(tree_), key=len)), {idx:1})
                jet_groomed=nx.get_node_attributes(tree_groomed,'pseudojet')[1]
                
                # add jet label to the groomed jet tree:
                tree_groomed.add_node(1,jet_label=r'groomed jet ($p_t$={} GeV)'.format(round(jet_groomed.pt,1)))
                draw_trees([tree,tree_groomed],jet_label=True)
                                
````
